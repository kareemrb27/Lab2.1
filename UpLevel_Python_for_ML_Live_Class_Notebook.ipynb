### Copy vs View

```python
# Copy
#  - New array is independent and does not share any memory with the original array
#  - Modifying the new array will not affect the original array
#  - The two arrays are different Python objects that point to different data in memory
```

```python
# Define new array
arr = np.array([1, 2, 3, 4, 5])
copied_arr = arr.copy()
```

```python
print(arr)
print(id(arr))
print()
print(copied_arr)
print(id(copied_arr))
```

```python
# Change original array
arr[0] = 42
```

```python
print('Original Array:', arr)
print()
print('Copied Array:', copied_arr)
```

```python
# Change copied array
copied_arr[-1] = 100
```

```python
print('Original Array:', arr)
print()
print('Copied Array:', copied_arr)
```

```python
# View
#  - The new array shares data with the original array
#  - Modifying the new array will affect the original array
#  - The arrays are different Python objects that point to same data in memory
```

```python
arr2 = np.array([1, 2, 3, 4, 5])
viewed_array = arr2.view()
```

```python
print(arr2)
print(id(arr2))
print()
print(viewed_array)
print(id(viewed_array))
```

```python
# Change original array
arr2[0] = 42
```

```python
print(arr2)
print()
print(viewed_array)
```

```python
# Change viewed array
viewed_array[1] = 80
```

```python
print(arr2)
print()
print(viewed_array)
```

### Filtering

```python
# Filter NumPy array
arr = np.array([41, 42, 43, 44])
print(arr)
```

```python
# Keep even numbers

# Option 1 - Manual filter
filter_arr = [False, True, False, True]

print(arr[filter_arr])
```

```python
# Option 2 - Create filter
filter_arr = []

for x in arr:
    if x % 2 == 0:
        filter_arr.append(True)
    else:
        filter_arr.append(False)

# print(filter_arr)
print(arr[filter_arr])
```

```python
# Option 3 - Shorter version
filter_arr = (arr % 2 == 0)

# print(filter_arr)
print(arr[filter_arr])
```

```python
# 2d example
arr = np.array([[1, 2, 3], [4, 5, 6]])

print('Original Array:\n\n', arr)
print()

filter_arr = (arr % 2 == 0)

print('\nFilters:\n\n', filter_arr)
print()

print('\nFiltered Array:\n\n', arr[filter_arr])
```

### Broadcasting

```python
array1 = [1, 2, 3]
array2 = [[1], [2], [3]]
```

```python
# Here, NumPy automatically broadcasts the size of a 1-D array array1 to perform element-wise addition with a 2-D array array2.

res = array1 + array2
```

```python
# create 1-D array
import numpy as np
array1 = np.array([1, 2, 3])

# create 2-D array
array2 = np.array([[1], [2], [3]])

# add arrays of different dimension
# size of array1 expands to match with array2
sum = array1 + array2

print(sum)
```

Numpy automatically expands the size of 1-D array array1 to match with the size of 2-D array array2.

Then, the element-wise addition is performed between two 2-D arrays.

Broadcasting with Scalars

```python
# 1-D array
array1 = np.array([1, 2, 3])

# scalar
number = 5

# add scalar and 1-D array
sum = array1 + number

print(sum)

# In this example, NumPy automatically expands the scalar number to an 1-D array and then performs the element-wise addition.
```

### Replace the missing value

```python
df = pd.DataFrame({
    'account_length': [30, 20, 20, np.nan, 30],
    'gender': ['F', 'M', 'F', np.nan, 'F'],
    'education': ['high school', 'high school','high school','high school',np.nan]
})

df
```

```python
# replace the missing value with average value of the column 'account_length'
value_to_replace = df['account_length'].mean()

value_to_replace

# 30 + 20 + 20 + 30 = 100
# 100 / 4 = 25
```

```python
# fill the msising value with the value_to_replace
df['account_length'].fillna(value_to_replace)
```

```python
df
```

```python
df['account_length'].fillna(value_to_replace, inplace = True)

df
```

```python
# replace the missing value with the most frequent value in column 'gender'
value_to_replace2 = df['gender'].mode()[0]

# value_to_replace = df['account_length'].mean()

value_to_replace2
```

```python
value_to_replace2 = df['gender'].mode()
# it returns a Series
# [0] access the first element of the Series

value_to_replace2
```

```python
df['gender'].fillna(value_to_replace2, inplace = True)

df
```



### Data Analysis : groupby, agg

```python
df = pd.DataFrame({
    'product_category': ['baby', 'book', 'book', 'beauty', 'baby'],
    'product_price': [19.99, 12.49, 22, 49, 16],
    'rating': [5, 3, 3, 2, 4]
})

df
```

### What is the average rating by product category?

```python
df.groupby('product_category')['rating'].mean()
```

```python
avg_rating_summary = df.groupby('product_category')['rating'].mean().reset_index()
# convert the grouped index back to columns

avg_rating_summary
```

```python
avg_rating_summary.rename(columns = {'rating': 'avg_rating'}, inplace = True)

avg_rating_summary
```

### What is the average product price and total quantity of each product category?

```python
df = pd.DataFrame({
    'order_id': ['4235', '2342', '1234', '5325', '1342'],
    'product_category': ['baby', 'beauty', 'baby', 'beauty', 'beauty'],
    'product': ['diaper', 'eye liner', 'diaper', 'eye shadow', 'face mask'],
    'product_price': [19.99, 12.49, 22, 49, 16],
    'quantity': [5, 3, 3, 2, 4]
})

df
```

```python
# What is the average product price and total quantity of each product category?
product_summary = df.groupby('product_category').agg({'product_price':'mean', 'quantity' : 'sum'}).reset_index()

product_summary
```

```python
product_summary.rename(columns = {'product_price': 'avg_product_price', 'quantity' :'total_quantity'}, inplace = True)

product_summary
```

### Merging

```python
# Merge based on the 'product_id' column
merged_df = product_df.merge(review_df, on = 'product_id', how = 'inner')

merged_df

# 3 rows?
# by default: merge function defaults to inner join
```
