# Introduction to Gradient Descent





Gradient Descent is a fundamental optimization algorithm used to minimize the cost function of a machine learning model. It plays a key role in adjusting model parameters iteratively to find the optimal solution. The core idea is to compute the gradient of the cost function with respect to the model parameters and update them in the opposite direction of the gradient, moving towards the minimum.

<center><img src="https://miro.medium.com/v2/resize:fit:1200/format:webp/0*7VyOHNGxbSvHoKAP.jpg" width=500/></center>

### How Gradient Descent Works:
- **Initialize Parameters:** Start by assigning random values to the model parameters.
- **Compute the Cost Function:** Evaluate the cost function with the current parameters.
- **Calculate the Gradient:** Find the gradient of the cost function with respect to each parameter, indicating the direction of steepest ascent.
- **Update Parameters:** Adjust each parameter by subtracting the product of the gradient and a predefined learning rate.
- **Repeat:** Continue steps 2-4 until the cost function converges to a minimum, signaling optimal parameters.

## Section 1: Gradient Descent with Fixed Learning Rate

Let's start by implementing gradient descent with a fixed learning rate in Python.


We will use the following cost function:

$
f(x) = (x - 5)^2 + 20
$


Let us the plot the function first!


1. **Step 1:** Define the mathematical function $f(x) = (x - 5)^2 + 20$ in the code. This function calculates the value for any input $x$.
  
2. **Step 2:** Create an array of 1000 equally spaced values of $x$ ranging from -10 to 20 using the `np.linspace()` function.

3. **Step 3:** Use the defined function $f(x)$ to compute the corresponding $y$ values for each value of $x$.

4. **Step 4:** Plot the computed $x$ and $y$ values using `plt.plot(x, y)`, which will generate a line plot of the function $f(x)$ versus $x$.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Create an array of x values
x = np.linspace(-10, 20, 1000)

# Evaluate f(x) for each value of x
y = f(x)


# Create a figure with specific dimensions (width, height)
plt.figure(figsize=(4, 4))

# Plot f(x) versus x
plt.plot(x, y)

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) versus x')

# Display the plot
plt.show()
```

### Observation:
- It appears to be a simple quadratic function with a minimum at x = 5.




We will use gradient descent to find the minimum of this function.


1. **Define the cost function**: $f(x) = (x - 5)^2 + 20$ and its gradient: $ \text{gradient}(x) = 2 \times (x - 5) $.

2. **Initialize**: Set the starting point $x$ using `np.random.randn()`, the learning rate to 0.1, and max iterations to 100.

3. **Track cost**: Create a list `cost_values` to store the cost at each iteration.

4. **Optimization Loop**: For 100 iterations:
   - Calculate the cost using $f(x)$).
   - Compute the gradient at $x$.
   - Update $x$ using $ x = x - \text{learning rate} \times \text{grad} $.
   - Track the cost and print progress every 10 iterations.

5. **Plot the results**: Visualize how the cost decreases over the iterations.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return (x - 5) ** 2 + 20

# Define the gradient of the cost function
def gradient(x):
    return 2 * (x - 5)

# Initialize the model parameter with a random value
x = np.random.randn()


# Set the learning rate
learning_rate = 0.1

# Set the maximum number of iterations
max_iterations = 100

# Initialize the list to store the cost values
cost_values = []

# Iterate until convergence or max iterations
for i in range(max_iterations):
    # Calculate the cost and gradient at the current parameter
    cost = cost_function(x)
    grad = gradient(x)

    # Print the progress
    if i % 10 == 0:
        print("Iteration {}: x = {}, cost = {}".format(i, x, cost))

    # Update the parameter
    x = x - learning_rate * grad

    # Store the cost value
    cost_values.append(cost)



# Plot the cost values
plt.figure(figsize=(4, 4))
plt.plot(cost_values)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

### Interpretation:

- At the beginning (iteration 0), the cost is relatively high, around 50. This indicates that the initial random value of $x$ is far from the optimal value that minimizes the cost function.

- In the first few iterations (roughly 0 to 20), the cost decreases significantly. This steep drop shows that the gradient descent algorithm is making large improvements in reducing the cost at the start, as the gradient is likely larger when the initial $x$ is farther from the optimal point.

- After about 20 iterations, the cost flattens out and approaches a value of around 20. This flattening indicates that the algorithm is converging to the minimum of the cost function, and subsequent updates to $x$ lead to smaller changes in the cost.

- The cost stabilizes around 20, which is the minimum value for the given cost function. This shows that the optimization process successfully reached the minimum.

The plot demonstrates a successful gradient descent process, with a sharp initial decrease in cost followed by gradual convergence to the minimum. This is typical behavior for gradient-based optimization algorithms when applied to convex functions like the quadratic one in this example.

Let's observe how x changes after each iteration of Gradient Descent!

Heres a more concise version of the steps:

1. **Define the function and gradient**: Use $f(x) = (x - 5)^2 + 20$ and its gradient $ \text{grad}_f(x) = 2 \times (x - 5) $.

2. **Gradient Descent Algorithm**: Implement `gradient_descent()` that iteratively updates $ x $ by subtracting the gradient scaled by the learning rate over `num_iterations`, storing all $ x$-values.

3. **Set Hyperparameters**: Choose learning rate $ \alpha = 0.1$, number of iterations `5`, and initial point $ x_{\text{init}} = 15 $.

4. **Run Gradient Descent**: Call `gradient_descent()` to get the sequence of $x$ -values showing the steps of the algorithm.

5. **Prepare Plot**: Generate points for $f(x) $ and plot the cost function curve.

6. **Plot Steps**: Plot the movement between consecutive \$x$-values from `x_list`, labeling each step with its cost.


This helps to visualize the steps of gradient descent on a quadratic cost function, showing how the algorithm iteratively minimizes the cost.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Define the gradient of f(x)
def grad_f(x):
    return 2 * (x - 5)

# Define the gradient descent algorithm
def gradient_descent(f, grad_f, x_init, alpha, num_iterations):
    x = x_init
    x_list = [x]
    for i in range(num_iterations):
        x = x - alpha * grad_f(x)
        x_list.append(x)
    return x_list

# Set the hyperparameters
alpha = 0.1
num_iterations = 5

# Run gradient descent on f(x)
x_init = 15
x_list = gradient_descent(f, grad_f, x_init, alpha, num_iterations)

# Create an array of x values
x = np.linspace(-2, 15, 100)

# Evaluate f(x) for each value of x
y = f(x)

# Plot f(x) versus x
plt.figure(figsize=(4, 4))
plt.plot(x, y)

# Plot the gradient descent iterations
for i in range(len(x_list) - 1):
    x1 = x_list[i]
    y1 = f(x1)
    x2 = x_list[i + 1]
    y2 = f(x2)
    plt.plot([x1, x2], [y1, y2], 'ro-')
    plt.text(x1, y1 + 0.5, round(y1, 2))

# Label the final cost value
x_final = x_list[-1]
y_final = f(x_final)
plt.text(x_final, y_final + 0.5, round(y_final, 2))

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient descent iterations on f(x)')

# Display the plot
plt.show()
```

## Section 2: Impact of learning rate (alpha) on convergence

The learning rate (`alpha`) in gradient descent controls how large the step sizes are during the optimization process. Its impact on convergence is crucial:

1. **Small Learning Rate (Low alpha):**
   - **Effect:** Slow convergence.
   - **Explanation:** If `alpha` is too small, the algorithm will take tiny steps towards the minimum, requiring more iterations to converge. While this ensures stability, it might take too long to reach the solution.

2. **Optimal Learning Rate:**
   - **Effect:** Fast and efficient convergence.
   - **Explanation:** When `alpha` is just right, the algorithm will converge quickly to the minimum without overshooting or slowing down too much.

3. **Large Learning Rate (High alpha):**
   - **Effect:** Divergence or oscillation.
   - **Explanation:** If `alpha` is too large, the algorithm may take steps that are too big, causing it to overshoot the minimum. This can lead to oscillations around the minimum or even cause the algorithm to diverge, failing to converge at all.

Thus, choosing an appropriate learning rate is critical for ensuring the gradient descent algorithm converges efficiently and reliably.

Let's check the gradient descent plots for different learning rates!


1. **Define the cost function and its gradient:** `cost_function(x)` computes the cost, and `gradient(x)` calculates the gradient.

2. **Set hyperparameters:** Define `max_iterations = 100` and different `learning_rates = [0.001, 0.01, 0.1, 1]` to test multiple learning rates.

3. **Initialize storage:** Create an empty list `cost_values` to store cost values for each learning rate.

4. **Iterate over learning rates:** For each learning rate:
   - Initialize a random starting value for `x`.
   - Perform gradient descent for `max_iterations`, updating `x` based on the learning rate and gradient.
   - Store the computed cost for each iteration.

5. **Plot the cost:** After the iterations, plot the cost values over iterations for each learning rate.

6. **Finalize plot:** Add legend, labels, and display the plot to compare how the different learning rates affect convergence.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return (x - 5) ** 2 + 20

# Define the gradient of the cost function
def gradient(x):
    return 2 * (x - 5)

# Set the maximum number of iterations
max_iterations = 100

# Initialize the list to store the cost values for different learning rates
learning_rates = [0.001, 0.01, 0.1, 1]
cost_values = [[] for _ in range(len(learning_rates))]

plt.figure(figsize=(4, 4))
# Iterate over different learning rates
for j, learning_rate in enumerate(learning_rates):
    # Initialize the model parameter with a random value

    np.random.seed(42)
    x = np.random.randn()

    # Iterate until convergence or max iterations
    for i in range(max_iterations):
        # Calculate the cost and gradient at the current parameter
        cost = cost_function(x)
        grad = gradient(x)

        # Update the parameter
        x = x - learning_rate * grad

        # Store the cost value
        cost_values[j].append(cost)

    # Plot the cost values for this learning rate
    plt.plot(cost_values[j], label="learning rate = {}".format(learning_rate))

# Add legend and labels to the plot
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

### Interpretation:

- Lower learning rates (0.001, 0.01) lead to slow but stable convergence.
- An optimal learning rate (0.1) results in the fastest and most efficient convergence.
- Higher learning rates (1) cause the algorithm to diverge, preventing convergence to the minimum.

## Section 3: Gradient Descent Optimization

### Types of Gradient Descent:

1. **Batch Gradient Descent (BGD):**
   - In **Batch Gradient Descent**, the model updates the weights by computing the gradient of the cost function using the **entire dataset** at each iteration. The entire dataset is passed through the model to calculate the cost and gradient.
   - **Pros:**
     - Provides a more stable and smoother convergence.
     - Fewer noisy updates because the gradient is calculated over the whole dataset.
   - **Cons:**
     - Computationally expensive for large datasets since each iteration requires processing the entire dataset.
     - Slower updates compared to other methods.
   
2. **Mini-Batch Gradient Descent (MBGD):**
   -  **Mini-Batch Gradient Descent** splits the training dataset into smaller batches (called mini-batches), and the model updates the weights by computing the gradient using each mini-batch. This method combines elements of both Batch Gradient Descent and Stochastic Gradient Descent.
   - **Pros:**
     - Faster than batch gradient descent because it processes smaller batches.
     - More stable than stochastic gradient descent due to averaging the gradients over the mini-batches.
     - Efficient on large datasets because it takes advantage of vectorized operations.
   - **Cons:**
     - Requires tuning the batch size, which can affect performance and convergence.
   
3. **Stochastic Gradient Descent (SGD):**
   -  In **Stochastic Gradient Descent**, the model updates the weights by computing the gradient of the cost function using **one training example** at each iteration. The model processes one training sample at a time, which leads to quicker, but noisier, updates.
   - **Pros:**
     - Faster updates as it processes one sample at a time, making it more efficient for large datasets.
     - Can escape local minima because of its noisy updates, helping the model to potentially find a better global minimum.
   - **Cons:**
     - Noisy and may fluctuate around the minimum, making it harder to converge.
     - Requires more iterations to converge compared to batch gradient descent.

Let's implement **Batch Gradient Descent (BGD)**, **Mini-Batch Gradient Descent (MBGD)**, and **Stochastic Gradient Descent (SGD)** for logistic regression on the **Wine dataset** to compare how each method optimizes the model's weights.

```python
# Import the required libraries
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# Load the Wine dataset
wine = load_wine()
X = wine.data
y = (wine.target != 0) * 1  # Convert to binary classification (Class 0 vs. all)

# Standardize the dataset (important for gradient descent)
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add intercept term (bias) to X
X_train_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]  # Add a column of ones for intercept
X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]     # Same for test set

# Sigmoid function for logistic regression
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic regression cost function (binary cross-entropy)
def compute_cost(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

# Gradient of the cost function
def compute_gradient(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    gradient = (1/m) * X.T.dot(h - y)
    return gradient

# Batch Gradient Descent
def batch_gradient_descent(X, y, theta, alpha, iterations):
    # Initialize an empty list to store the cost at each iteration
    cost_history = []

    # Loop over the number of iterations
    for i in range(iterations):
        # Compute the gradient of the cost function with respect to theta
        gradient = compute_gradient(X, y, theta)

        # Update the parameters (theta) by taking a step in the direction of the negative gradient
        # alpha is the learning rate that controls the step size
        theta -= alpha * gradient

        # Calculate the cost with the updated theta and append it to the cost history list
        cost_history.append(compute_cost(X, y, theta))

    # Return the final parameters (theta) and the history of the cost function over the iterations
    return theta, cost_history


# Mini-Batch Gradient Descent
def mini_batch_gradient_descent(X, y, theta, alpha, iterations, batch_size):
    # Get the number of training examples (m)
    m = len(y)
    # Initialize an empty list to store the cost at each iteration
    cost_history = []

    # Loop over the number of iterations
    for i in range(iterations):
        # Shuffle the dataset at the start of each iteration to randomize the mini-batches
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        # Loop over the dataset in mini-batches
        for j in range(0, m, batch_size):
            # Select the mini-batch from the shuffled data
            X_mini_batch = X_shuffled[j:j+batch_size]
            y_mini_batch = y_shuffled[j:j+batch_size]

            # Compute the gradient using the mini-batch
            gradient = compute_gradient(X_mini_batch, y_mini_batch, theta)

            # Update the parameters (theta) using the computed gradient and learning rate (alpha)
            theta -= alpha * gradient

        # Calculate the cost with the updated theta using the entire dataset
        cost_history.append(compute_cost(X, y, theta))

    # Return the final parameters (theta) and the history of the cost function over the iterations
    return theta, cost_history


# Stochastic Gradient Descent
def stochastic_gradient_descent(X, y, theta, alpha, iterations):
    # Get the number of training examples (m)
    m = len(y)
    # Initialize an empty list to store the cost at each iteration
    cost_history = []

    # Loop over the number of iterations
    for i in range(iterations):
        # Loop over each training example in the dataset
        for j in range(m):
            # Randomly pick one training example at a time (stochastic)
            rand_index = np.random.randint(0, m)
            X_i = X[rand_index, :].reshape(1, X.shape[1])  # Select the random sample X
            y_i = y[rand_index].reshape(1)  # Select the corresponding label y

            # Compute the gradient using the single random sample
            gradient = compute_gradient(X_i, y_i, theta)

            # Update the parameters (theta) using the computed gradient and learning rate (alpha)
            theta -= alpha * gradient

        # Calculate the cost with the updated theta using the entire dataset
        cost_history.append(compute_cost(X, y, theta))

    # Return the final parameters (theta) and the history of the cost function over the iterations
    return theta, cost_history


# Prepare initial theta (weights)
n = X_train_b.shape[1]
theta_initial = np.zeros(n)

# Hyperparameters
alpha = 0.01
iterations = 1000
batch_size = 32  # For mini-batch

# ----------- Batch Gradient Descent -----------
theta_bgd, cost_history_bgd = batch_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations)

# ----------- Mini-Batch Gradient Descent -----------
theta_mbgd, cost_history_mbgd = mini_batch_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations, batch_size)

# ----------- Stochastic Gradient Descent -----------
theta_sgd, cost_history_sgd = stochastic_gradient_descent(X_train_b, y_train, theta_initial.copy(), alpha, iterations)

# Plot the cost history for each method
plt.figure(figsize=(6, 6))
plt.plot(range(iterations), cost_history_bgd, label='Batch Gradient Descent', color='r')
plt.plot(range(iterations), cost_history_mbgd, label='Mini-Batch Gradient Descent', color='g')
plt.plot(range(iterations), cost_history_sgd, label='Stochastic Gradient Descent', color='b')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost History for Different Gradient Descent Methods')
plt.legend()
plt.show()

# Predict function for logistic regression
def predict(X, theta):
    return np.round(sigmoid(X.dot(theta)))

# Evaluate all three models
y_pred_bgd = predict(X_test_b, theta_bgd)
y_pred_mbgd = predict(X_test_b, theta_mbgd)
y_pred_sgd = predict(X_test_b, theta_sgd)

# Print classification reports
print("Classification Report for Batch Gradient Descent:\n")
print(classification_report(y_test, y_pred_bgd))

print("\nClassification Report for Mini-Batch Gradient Descent:\n")
print(classification_report(y_test, y_pred_mbgd))

print("\nClassification Report for Stochastic Gradient Descent:\n")
print(classification_report(y_test, y_pred_sgd))
```

### Interpretation
- Stochastic Gradient Descent (blue) converges the fastest but shows initial fluctuations due to noisy updates.
- Mini-Batch Gradient Descent (green) converges more quickly than Batch Gradient Descent while maintaining stability.
- Batch Gradient Descent (red) converges steadily but more slowly than the other two methods.
- Stochastic Gradient Descent reaches the lowest cost value, while Batch Gradient Descent takes the longest to reduce the cost.
- Mini-Batch Gradient Descent provides a good trade-off between convergence speed and stability.
