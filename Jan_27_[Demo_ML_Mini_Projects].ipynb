```python
# Decontract text
# We've --> We have


# Decontracting text involves expanding contractions into their full forms.
def decontract_text(text):

    # decontract text

    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"we’ve", "we have", text)
    text = re.sub(r"can\'t", "can not", text)

    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)

    return text
```

```python
import re
decontract_text(text.lower())
```

```python
text = text.lower().replace('\n', ' ')
text.lower().replace('\u2060,', ' ')
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
sw # Generic stopwords
```

```python
len(sw)
```

```python
sw.extend(['?', ',', "'", '"', '.'])
sw
```

```python
for i in word_tokenize(text):
  if i not in sw:
    print(i)# non-stopwords
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
ss.stem('increases')
```

```python
sw.append("’")
for i in word_tokenize(text):
  if i not in sw:
    print(i, ss.stem(i))# non-stopwords
```

```python
sw.append("’")
final_words = []
for i in word_tokenize(text):
  if i not in sw:
    final_words.append(ss.stem(i))# non-stopwords
```

```python
' '.join(final_words)
```

```python
doc = ' '.join(final_words)
cv.fit([doc])
```

```python
cv.vocabulary_
```

```python
sorted(cv.vocabulary_)
```

```python
pd.DataFrame(cv.transform([doc]).toarray(), columns = sorted(cv.vocabulary_))
```
